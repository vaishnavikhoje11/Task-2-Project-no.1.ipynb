```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("retail_sales_dataset (1).csv")
```

```python
print(df.head())
print(df.info())
```

```python
total_revenue = df['Total Amount'].sum()
print("Total Revenue:", total_revenue)
```

```python
avg_order_value = df.groupby('Transaction ID')['Total Amount'].sum().mean()
print("Average Order Value:", avg_order_value)
```

```python
top_products = df.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False)

print(top_products)
```

```python
monthly_sales = df.groupby('Month')['Total Amount'].sum()
print(monthly_sales)
```

```python
plt.figure(figsize=(8,5))
monthly_sales.plot(kind='line', marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.figure(figsize=(8,5))
top_products.plot(kind='bar')
plt.title("Top Product Categories")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.show()
```

```python
gender_sales = df.groupby('Gender')['Total Amount'].sum()

plt.figure(figsize=(6,4))
gender_sales.plot(kind='bar')
plt.title("Sales by Gender")
plt.xlabel("Gender")
plt.ylabel("Revenue")
plt.show()
```

```python
print("\n----- SALES ANALYSIS SUMMARY -----")

print("Total Revenue:", total_revenue)

print("Average Order Value:", avg_order_value)

print("Top Product Category:", top_products.idxmax())

print("\nTop Categories by Revenue:\n")
print(top_products)
```
